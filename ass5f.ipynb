```python
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
```

```python
corpus = input("Enter your text corpus: ")
```

```python
corpus = re.sub('[^A-Za-z ]+', ' ', corpus)
```

```python
corpus = corpus.lower()
```

```python
words = word_tokenize(corpus)
```

```python
words = [w for w in words if w not in stopwords.words('english')]
```

```python
print("\nCleaned words:\n", words)
```

```python
window_size = 2
data = []
```

```python
for i in range(window_size, len(words)-window_size):
    context = [words[i-2], words[i-1], words[i+1], words[i+2]]
    target = words[i]
    data.append((context, target))
```

```python
print("\nContext -> Target word pairs:")
for context, target in data:
    print(context, "->", target)
```

```python
sentences = [words]
```

```python
model = Word2Vec(sentences, vector_size=50, window=2, min_count=1, sg=0)
```

```python
print("\nVocabulary:", list(model.wv.index_to_key))
```

```python
print("\nEnter 4 context words (2 before + 2 after the target word):")
context_words = []
for i in range(4):
    w = input(f"Word {i+1}: ").lower()
    if w not in model.wv.key_to_index:
        print(f"'{w}' not in vocabulary! Try again.")
        exit()
    context_words.append(w)
```

```python
predicted = model.predict_output_word(context_words)
if predicted:
    print("\nPredicted middle word:", predicted[0][0])
else:
    print("\nCould not predict the middle word. Try different context words.")
predicted
```
